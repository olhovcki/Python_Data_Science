**Работа с данными в Pandas. 
Задание 1.** Импортируйте библиотеку Pandas и дайте ей псевдоним pd. Создайте датафрейм authors со столбцами author_id и author_name, в которых соответственно содержатся данные: [1, 2, 3] и ['Тургенев', 'Чехов', 'Островский'].


```python
import pandas as pd
```

```python
a = {
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
}

authors = pd.DataFrame(a)
authors
```

```python
a = {
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price":[450, 300, 350, 500, 450, 370, 290]
}

books = pd.DataFrame(a)
books
```

**Задание 2.**
Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.



```python
authors_price = pd.merge(authors, books, on='author_id', how='outer')
authors_price
```

**Задание 3.**
Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.


```python
top5 = authors_price.nlargest(5, "price")
top5
```

**Задание 4.**
Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:


```python
authors_stat = authors.copy()
authors_stat
```

```python
tmp = authors_price.groupby("author_id").agg({"price": "max"})
tmp.columns = ['max_price']
tmp

authors_stat = pd.merge(authors_stat, tmp, on='author_id', how='inner')
authors_stat
```

```python
tmp = authors_price.groupby("author_id").agg({"price": "min"})
tmp.columns = ['min_price']
tmp

authors_stat = pd.merge(authors_stat, tmp, on='author_id', how='inner')
authors_stat
```

```python
tmp = authors_price.groupby("author_id").agg({"price": "mean"})
tmp.columns = ['mean_price']
tmp

authors_stat = pd.merge(authors_stat, tmp, on='author_id', how='inner')
authors_stat
```

```python
authors_stat.drop(['author_id'], axis='columns', inplace=True)
authors_stat
```
